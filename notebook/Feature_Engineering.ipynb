```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/DataCoSupplyChainDataset.csv", encoding="latin1")

df.head()
```

```python
df = df.drop_duplicates()
```

```python
important_cols = [
    "Type",
    "Days for shipping (real)",
    "Days for shipment (scheduled)",
    "Benefit per order",
    "Sales per customer",
    "Delivery Status",
    "Late_delivery_risk",
    "Category Name",
    "Customer Segment",
    "Market",
    "Order Region",
    "Order State",
    "Order Status",
    "Order Item Discount",
    "Order Item Discount Rate",
    "Order Item Product Price",
    "Order Item Profit Ratio",
    "Order Item Quantity",
    "Sales",
    "Order Item Total",
    "Order Profit Per Order",
    "Product Price",
    "Shipping Mode"
]

df = df[important_cols]
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df["Delay_Gap"] = df["Days for shipping (real)"] - df["Days for shipment (scheduled)"]

df["Shipping_Pressure_Index"] = (
    df["Order Item Quantity"] / (df["Days for shipment (scheduled)"] + 1)
)

df["Order_Complexity_Score"] = (
    df["Order Item Quantity"] * df["Product Price"]
)

df["Profit_Sensitivity"] = (
    df["Order Profit Per Order"] / (df["Sales"] + 1)
)

df["Discount_Impact"] = (
    df["Order Item Discount Rate"] * df["Order Item Product Price"]
)
```

```python
df["Risk_Category"] = pd.cut(
    df["Late_delivery_risk"],
    bins=[-1, 0, 1],
    labels=["Low Risk", "High Risk"]
)
```

```python
df.to_csv("../data/processed/cleaned_data.csv", index=False)

print("Cleaned data saved successfully!")
print(df.shape)
```
